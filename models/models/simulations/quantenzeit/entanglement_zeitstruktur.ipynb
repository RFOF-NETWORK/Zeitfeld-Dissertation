# Simulation: Entanglement-Struktur der Quantenzeit

Dieses Notebook simuliert eine vereinfachte Entanglement-basierte
Zeitstruktur nach dem Page–Wootters-Formalismus. Ziel ist es, die
emergente Zeitrichtung aus Korrelationen zwischen einem "Uhrsystem"
und einem "physikalischen System" sichtbar zu machen.

Diese Struktur bildet die fundamentale Schicht des integrierten
    Zeitfeldmodells.

```python
import numpy as np
import matplotlib.pyplot as plt

# Dimension des Hilbertraums
N = 2000

# Zeitbasis für die Uhr
t = np.linspace(0, 10, N)

# Uhrzustand: Gauß-Paket
sigma = 0.4
clock_state = np.exp(-(t - 5)**2 / (2 * sigma**2))
clock_state /= np.linalg.norm(clock_state)
```

## Systemzustand: Energieeigenbasis

Wir modellieren ein einfaches Zweiniveau-System, dessen Dynamik durch die
Uhr getriggert wird.

```python
# Energieniveaus
E0 = 0.0
E1 = 1.0

# Systemzustände
psi0 = np.ones(N) / np.sqrt(N)
psi1 = np.exp(-1j * E1 * t)

# Gesamtzustand (entangled)
Psi = clock_state * psi1
```

## Emergenz der Zeitrichtung

Die effektive Zeitrichtung ergibt sich aus der Phasenentwicklung des
Systemzustands relativ zur Uhr.

```python
# Effektive Zeitrate = Phasenderivative
phase = np.unwrap(np.angle(Psi))
time_rate = np.gradient(phase, t)
```

## Plot: Uhrzustand, Systemphase und Zeitrate

```python
plt.figure(figsize=(14,6))

plt.subplot(1,3,1)
plt.plot(t, clock_state)
plt.title('Uhrzustand |Clock⟩')
plt.xlabel('t')
plt.ylabel('Amplitude')

plt.subplot(1,3,2)
plt.plot(t, phase)
plt.title('Systemphase')
plt.xlabel('t')
plt.ylabel('Phase')

plt.subplot(1,3,3)
plt.plot(t, time_rate)
plt.title('Emergente Zeitrate dτ/dt')
plt.xlabel('t')
plt.ylabel('Zeitfluss')

plt.tight_layout()
plt.show()
```

## Ergebnis

Dieses Notebook zeigt:

- die Entanglement-basierte Kopplung zwischen Uhr und System,
- die Phasenentwicklung des Systemzustands,
- die daraus emergente Zeitrichtung,
- die effektive Zeitrate als Ableitung der Phase.

Diese Struktur fließt direkt in die quantenphysikalische Schicht des
    Zeitfeldtensors ein:

$$ \mathcal{T}^{(q)}_{00} = \langle \Psi | \hat{H} | \Psi \rangle $$

und definiert die fundamentale Zeitrichtung des gesamten Modells.
