# Analyse: Zyklusanalyse der Zeitfeld-Komponenten

Dieses Notebook untersucht die periodischen Strukturen der verschiedenen
Zyklen im integrierten Zeitfeldmodell. Ziel ist es, dominante Frequenzen,
Resonanzen und Schwebungen sichtbar zu machen.

## 1. Import der Bibliotheken

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.fft import rfft, rfftfreq
```

## 2. Beispielhafte Zyklusdefinitionen

Wir definieren exemplarische Frequenzen für planetare, solare und galaktische Zyklen.

```python
# Frequenzen in 1/Jahr
omega_p1 = 1/1.0     # Erde
omega_p2 = 1/11.86   # Jupiter
omega_s1 = 1/11      # Schwabe-Zyklus
omega_g1 = 1/240e6   # galaktische Modulation

# Zeitachse
t = np.linspace(0, 500, 200000)
```

## 3. Konstruktion der Signale

```python
A1, A2, A3, A4 = 1.0, 0.7, 0.4, 0.2

sig_p1 = A1 * np.cos(2*np.pi*omega_p1 * t)
sig_p2 = A2 * np.cos(2*np.pi*omega_p2 * t)
sig_s1 = A3 * np.cos(2*np.pi*omega_s1 * t)
sig_g1 = A4 * np.cos(2*np.pi*omega_g1 * t)

signal = sig_p1 + sig_p2 + sig_s1 + sig_g1
```

## 4. Fourier-Analyse

Die Fourier-Transformation zeigt die dominanten Frequenzen und Resonanzen.

```python
freqs = rfftfreq(len(t), d=(t[1] - t[0]))
spectrum = np.abs(rfft(signal))
```

## 5. Peaks im Spektrum finden

```python
peaks, _ = find_peaks(spectrum, height=np.max(spectrum)*0.1)
dominant_freqs = freqs[peaks]
dominant_freqs[:10]
```

## 6. Plot: Signal und Spektrum

```python
plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
plt.plot(t[:20000], signal[:20000], linewidth=0.5)
plt.title('Zyklisches Gesamtsignal')
plt.xlabel('Zeit (Jahre)')
plt.ylabel('Amplitude')

plt.subplot(1,2,2)
plt.plot(freqs, spectrum)
plt.scatter(dominant_freqs, spectrum[peaks], color='red')
plt.title('Fourier-Spektrum')
plt.xlabel('Frequenz (1/Jahr)')
plt.ylabel('Spektrale Stärke')
plt.xlim(0, 0.2)

plt.tight_layout()
plt.show()
```

## 7. Interpretation

- Die Peaks entsprechen den dominanten Zyklen.
- Schwebungen entstehen durch Differenzen der Frequenzen.
- Langzeitmodulationen sind im niederfrequenten Bereich sichtbar.

Diese Analyse bildet die Grundlage für die Interferenz- und Tensoranalyse im integrierten Zeitfeldmodell.
