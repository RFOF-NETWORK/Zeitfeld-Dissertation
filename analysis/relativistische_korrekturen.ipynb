# Analyse: Relativistische Zeitkorrekturen

Dieses Notebook untersucht die relativistischen Beiträge zur beobachteten Zeitrate.
Im integrierten Zeitfeldmodell bildet die relativistische Schicht (r) die Grundlage
aller makroskopischen Zeitdilatationen.

## 1. Grundlagen der relativistischen Zeitdilatation

Die Zeitdilatation ergibt sich aus zwei Hauptquellen:

- **Gravitative Zeitdilatation** (Allgemeine Relativitätstheorie)
- **Kinematische Zeitdilatation** (Spezielle Relativitätstheorie)

Die allgemeine Formel lautet:

$$ d\tau = dt \sqrt{1 - \frac{v^2}{c^2} - \frac{2GM}{rc^2}} $$

```python
import numpy as np
import matplotlib.pyplot as plt

c = 299792458  # Lichtgeschwindigkeit (m/s)
G = 6.67430e-11
M_earth = 5.972e24
R_earth = 6.371e6
```

## 2. Gravitative Zeitdilatation

Für ein Objekt in Höhe $h$ über der Erdoberfläche gilt:

$$ \frac{d\tau}{dt} = \sqrt{1 - \frac{2GM}{(R + h)c^2}} $$

```python
h = np.linspace(0, 40000, 500)  # 0–40 km
factor_grav = np.sqrt(1 - (2 * G * M_earth) / ((R_earth + h) * c**2))
```

## 3. Kinematische Zeitdilatation

Für ein Objekt mit Geschwindigkeit $v$ gilt:

$$ \frac{d\tau}{dt} = \sqrt{1 - \frac{v^2}{c^2}} $$

```python
v = np.linspace(0, 8000, 500)  # 0–8 km/s
factor_kin = np.sqrt(1 - (v**2) / c**2)
```

## 4. Visualisierung der Zeitdilatation

```python
plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
plt.plot(h, factor_grav)
plt.title('Gravitative Zeitdilatation')
plt.xlabel('Höhe (m)')
plt.ylabel('dτ/dt')

plt.subplot(1,2,2)
plt.plot(v, factor_kin)
plt.title('Kinematische Zeitdilatation')
plt.xlabel('Geschwindigkeit (m/s)')
plt.ylabel('dτ/dt')

plt.tight_layout()
plt.show()
```

## 5. Bedeutung für das Zeitfeldmodell

Die relativistische Schicht liefert die makroskopische Grundstruktur der Zeitrate.
Sie koppelt im Zeitfeldtensor an:

- galaktische Potentiale,
- solare Modulationen,
- planetare Zyklen.

Die vollständige Zeitrate ergibt sich aus:

$$ \frac{d\tau}{dt} = f(T^{(q)}, T^{(r)}, T^{(g)}, T^{(s)}, T^{(p)}) $$
