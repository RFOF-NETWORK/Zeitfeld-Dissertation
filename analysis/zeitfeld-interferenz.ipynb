# Analyse: Interferenz der Zeitfeld-Zyklen

Dieses Notebook untersucht die Interferenzstrukturen zwischen den verschiedenen
Zyklen des integrierten Zeitfeldmodells:

- planetare Frequenzen
- solare Aktivitätszyklen
- galaktische Modulationen
- relativistische Gradienten

Ziel ist es, Schwebungen, Resonanzen und Langzeitmodulationen sichtbar zu machen.

## 1. Import der Bibliotheken

Die Analyse nutzt Fourier-Methoden, Glättungsverfahren und
Interferenzsimulationen.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.fft import rfft, rfftfreq

# Falls du eigene Utilities nutzt:
# from tools.utilities.plotting import plot_interference
# from tools.utilities.constants import SOLAR_CYCLE_YEARS
# from tools.utilities.astro_helpers import load_planetary_cycles
```

## 2. Laden der Zyklusdaten

Hier werden die Frequenzen aus den Datenmodellen geladen.
Die JSON-Dateien liegen unter:

`models/data_models/zyklusparameter.json`

```python
import json

with open('../models/data_models/zyklusparameter.json', 'r') as f:
    zyklusdaten = json.load(f)

zyklusdaten
```

## 3. Konstruktion der Zeitreihen

Aus den Frequenzen werden synthetische Zeitreihen erzeugt,
um Interferenzstrukturen sichtbar zu machen.

```python
def generate_cycle(freq, duration=5000, dt=0.1):
    t = np.arange(0, duration, dt)
    signal = np.sin(2 * np.pi * freq * t)
    return t, signal

# Beispiel: Planetare Frequenz
freq_planet = zyklusdaten['planetare_frequenzen']['jupiter']
t, s_planet = generate_cycle(freq_planet)
```

## 4. Interferenz zweier Zyklen

Die Interferenz zweier Frequenzen erzeugt Schwebungen.
Dies ist zentral für das Zeitfeldmodell.

```python
freq_solar = zyklusdaten['solare_zyklen']['schwabe']
_, s_solar = generate_cycle(freq_solar)

interferenz = s_planet + s_solar

plt.figure(figsize=(12,4))
plt.plot(t[:5000], interferenz[:5000])
plt.title('Interferenz: Planetar + Solar')
plt.xlabel('Zeit')
plt.ylabel('Amplitude')
plt.show()
```

## 5. Fourier-Analyse

Die Fourier-Analyse zeigt die dominanten Frequenzen und Schwebungsstrukturen.

```python
yf = rfft(interferenz)
xf = rfftfreq(len(interferenz), 0.1)

plt.figure(figsize=(12,4))
plt.plot(xf, np.abs(yf))
plt.title('Fourier-Spektrum der Interferenz')
plt.xlabel('Frequenz')
plt.ylabel('Amplitude')
plt.xlim(0, 0.01)
plt.show()
```

## 6. Interpretation

- Peaks im Spektrum entsprechen dominanten Zyklen.
- Schwebungsfrequenzen entstehen durch Differenzen der Hauptfrequenzen.
- Langzeitmodulationen sind sichtbar als sehr niedrige Frequenzen.

Diese Strukturen sind zentral für das integrierte Zeitfeldmodell.
